```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_temp = 'data/temperature.csv'
temperature = pd.read_csv(file_temp)
print(temperature.info())
```

## Moving average of temperature change
The application of a 5-year moving average clarifies the underlying trajectory of annual surface temperature changes in Ecuador from 1961 to 2022. While year-to-year fluctuations still occur, the smoothed line highlights a gradual and sustained warming trend, particularly evident from the early 1990s onward. Recent decades consistently exceed the long-run average, underscoring the cumulative effect of climate change on regional temperatures.

```python
# Moving average of temperature change in Ecuador
# Set the window size
window_size = 5

# Calculate the moving average
ecu_data['Moving Average'] = ecu_data['Temperature Change'].rolling(window=window_size).mean()

# Plot
plt.figure(figsize=(10, 4))
plt.title(f'{window_size}-Year Moving Average of Temperature Change in Ecuador')
plt.plot(ecu_data.index, ecu_data['Temperature Change'], label='Annual Temperature Change', alpha=0.5)
plt.plot(ecu_data.index, ecu_data['Moving Average'], label=f'{window_size}-Year Moving Average', color='crimson', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Temperature Change (C)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Regression analysis
The linear regression fitted to Ecuadors annual surface temperature changes from 1961 to 2022 reveals a statistically significant and robust warming trend. The slope of 0.0224C per year indicates that the country has been warming at a rate of approximately 0.22C per decade. With an R-squared value of 0.602, the model explains over 60% of the variance in annual temperature, suggesting a strong, time-linked component in the warming signal. The low p-value further confirms that this trend is unlikely to be the result of random variation.

```python
from scipy.stats import linregress

# Convert datetime index to numeric years
years = ecu_data.index.year
temps = ecu_data['Temperature Change']

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(years, temps)

# Key results
print(f'Slope: {slope:.4f} C/year')
print(f'Intercept: {intercept:.2f}')
print(f'R-squared: {r_value**2:.3f}')
print(f'P-value: {p_value:.4e}')

# Predicted temperatures using the regression line
reg_line = intercept + slope * years

plt.figure(figsize=(10, 4))
plt.title('Linear Regression of Temperature Change in Ecuador')
plt.plot(years, temps, label='Observed Temperature Change', alpha=0.6)
plt.plot(years, reg_line, color='crimson', linestyle='--', label='Trend Line')
plt.xlabel('Year')
plt.ylabel('Temperature Change (C)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Warmest and Coldest Years in Ecuador
Using deviations from the long-term mean, the analysis highlights the five warmest and five coldest years in Ecuador from 1961 to 2022. The warmest year on record is 2016, with a temperature anomaly over 1C above the national average. Other notably hot years include 2020, 2015, 1998, and 2019, several of which align with major El Nio events.

Conversely, the coldest years, including 1971, 1967, 1968, 1975, and 1974, cluster in earlier decades, reinforcing the shift toward consistently warmer conditions in recent years.

These flagged years serve as climate landmarks, helping frame Ecuadors warming trend in both global and local context.

```python
# Top warmest/coldest years
mean_change = ecu_data['Temperature Change'].mean()
differences = (ecu_data['Temperature Change'] - mean_change).sort_values(ascending=False)
print('Top 5 Warmest Years')
print(differences.head(5).round(3))
print('\n')
print('Top 5 Coldest Years')
print(differences.tail(5).sort_values().round(3))
```

## Relationship between carbon emission and temperature
To assess the relationship between atmospheric CO and surface temperature, two linear regression models were fitted using Ecuadors annual data from 1961 to 2022.

```python
filepath = 'data/carbon_last_modification.csv'
carbon_last = pd.read_csv(filepath, parse_dates=['Date'], index_col='Date')
print(carbon_last.info())
```

```python
# Add yearly carbon emission data
co2_measure_mean = carbon_last['Value'].resample('YE').mean()
co2_measure_median = carbon_last['Value'].resample('YE').median()
start_date = '1961-12-31'
end_date = '2022-12-31'
filtered_measure_mean = co2_measure_mean.loc[start_date:end_date]
filtered_measure_median = co2_measure_median.loc[start_date:end_date]
ecu_data['Average Carbon Emission'] = filtered_measure_mean
ecu_data['Median Carbon Emission'] = filtered_measure_median
print(ecu_data.info())
```

### Carbon Emissions vs. Average Temperature Change
With the mean CO concentration as predictor, the regression yielded a slope of 0.0133C per ppm, indicating that each unit rise in average CO corresponds to a 0.0133C increase in temperature. The model achieved an R of 0.590, meaning nearly 60% of the temperature variability can be explained by carbon buildup.

```python
# Predictor and response
x = ecu_data['Average Carbon Emission']
y = ecu_data['Temperature Change']

# Fit regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Print regression results
print(f'Slope: {slope:.4f} C per ppm')
print(f'Intercept: {intercept:.2f}')
print(f'R-squared: {r_value**2:.3f}')
print(f'P-value: {p_value:.4e}')

# Regression line for plotting
reg_line = intercept + slope * x

# Plot
plt.figure(figsize=(10, 4))
plt.title('Regression: Carbon Emissions vs Temperature Change')
plt.scatter(x, y, alpha=0.6, label='Observed Data')
plt.plot(x, reg_line, color='crimson', linestyle='--', label='Regression Line')
plt.xlabel('Average Annual CO Concentration (ppm)')
plt.ylabel('Annual Temperature Change (C)')
plt.legend()
plt.tight_layout()
plt.show()
```

### Carbon Emissions vs. Median Temperature Change
Using the median CO concentration produced a nearly identical result, with a slope of 0.0132C per ppm and an R of 0.588. This confirms the robustness of the relationship, regardless of whether mean or median CO is used to represent central tendency.

```python
# Predictor and response
x = ecu_data['Median Carbon Emission']
y = ecu_data['Temperature Change']

# Fit regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Print regression results
print(f'Slope: {slope:.4f} C per ppm')
print(f'Intercept: {intercept:.2f}')
print(f'R-squared: {r_value**2:.3f}')
print(f'P-value: {p_value:.4e}')

# Regression line for plotting
reg_line = intercept + slope * x

# Plot
plt.figure(figsize=(10, 4))
plt.title('Regression: Carbon Emissions vs Temperature Change')
plt.scatter(x, y, alpha=0.6, label='Observed Data')
plt.plot(x, reg_line, color='crimson', linestyle='--', label='Regression Line')
plt.xlabel('Median Annual CO Concentration (ppm)')
plt.ylabel('Annual Temperature Change (C)')
plt.legend()
plt.tight_layout()
plt.show()
```

In both cases, the results are statistically significant and strongly suggest that Ecuador's warming trend is closely tied to the broader rise in atmospheric CO, reinforcing the patterns uncovered in the time-series and decadal analyses.
